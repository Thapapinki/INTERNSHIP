# Bivariate Analysis

```python
# Analysising data from gender and the age
sns.swarmplot(x="1Gender of respondent", y="2 How old are you? ",data=custo_excel)
plt.xlabel('')
plt.ylabel('')
plt.show()
```

Observation - Female customers in the age group of 21-30 years do more shopping then male in the same age group. Customers(male and female) in the age group of 31-50 years are almost the same who pefer to do shopping. But after 50 years female customers are more than male customer.

```python
# Analysising data from gender and the city
sns.swarmplot(x="1Gender of respondent", y="3 Which city do you shop online from?",data=custo_excel)
plt.xlabel('')
plt.ylabel('')
plt.show()
```

Observations - 

1) In Bangalore maximum number of customer are female.

2) In Delhi male customer are more as compared to female customers.

3) In Bulandshahr and Moradabad not a single female do shopping. 

4) In Solan and Meerut only female customers do shopping and not a single man do shopping.

5) Overall female customers are more in numbers as compared to male customers in all the cities mentioned above.

```python
# Analysising data from gender and access the internet while shopping.
sns.swarmplot(x="1Gender of respondent",y="7 How do you access the internet while shopping on-line?",data=custo_excel)
plt.xlabel('')
plt.ylabel('')
plt.show()
```

Observation -

1)  Not a single female customer use dial-up option for shopping.

2) Male and female use equally mobile internet access for shopping.

3) Wifi access is used by maximum number of female customers than male.

```python
# Analysising data from gender and access the device for shopping.
sns.swarmplot(x="1Gender of respondent",y="8 Which device do you use to access the online shopping?",data=custo_excel)
plt.xlabel('')
plt.ylabel('')
plt.show()
```

Observation - 

1)  First device perference is given to Smartphone by the customers(male, female).

2)  Second device perference is given to Laptop by the customers(male, female).

3)  There are few cutomers who are using Desktop for the shopping.

4)  Tablet is used by only male customers. There is not a single female using the same device for shopping.

```python
# Analysising data from gender and shopping flexibility
plt.title('Shopping online is convenient and flexible',size=14)
sns.swarmplot(x="1Gender of respondent",y="32 Shopping online is convenient and flexible",data=custo_excel)
plt.xlabel('')
plt.ylabel('')
plt.show()
```

Observation - Maximum number of male and female customer strongly agrees that shopping online is convenient and flexible. There are few customers who are not effected at all from wheather shopping online is convenient andflexible.
